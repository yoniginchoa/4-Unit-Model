```python
import numpy as np 
data_lst = [np.loadtxt(f'./{_}.txt', delimiter=',') for _ in ['E', 'P', 'S', 'V']]
```

```python
import matplotlib.pyplot as plt
for _, data in zip(['E', 'PV', 'SST', 'VIP'], data_lst) : 
    plt.plot(data[:,0], data[:,1],'-o', label=f'{_}') 
plt.xlabel('Constrast (%)')
plt.ylabel('Mean Activity')
plt.legend()
```
